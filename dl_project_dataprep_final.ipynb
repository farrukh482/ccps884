```python
def shiftPricesByMonth(df):
  df.sort_values(by=['ratings_date'], inplace=True, ascending=True)
  df['month_later_price'] = df['price'].shift(-22)
  return df

df_ratings = df_ratings.groupby('ticker').apply(lambda x : shiftPricesByMonth(x)).drop(columns=['ticker']).reset_index().drop(columns=['level_1'])
df_ratings.sort_values(by=['ratings_date', 'ticker'], inplace=True, ascending=True)

for i,row in df_ratings.iterrows():
  df_ratings.at[i,'returns'] = (df_ratings.at[i, 'month_later_price'] - df_ratings.at[i, 'price']) / df_ratings.at[i, 'price']

df_ratings.head(25)
```
